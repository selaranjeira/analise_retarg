```python
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
```

#### tabela googleads

```python
df_google = pd.read_csv('/home/selaranjeira/Desktop/hurb_python/rawDataInstallInappEventsPur/dados/googleadwords.csv')
```

```python
df = df_google[['IsPrimaryAttribution', 'diff_days']]
```

```python
# df['diff_days'].groupby(df['diff_days']).value_counts()
sum_trues_PA_df = df.groupby(['diff_days']).agg({'IsPrimaryAttribution': 'sum'}) 
counting_df = df.groupby(['diff_days']).agg({'IsPrimaryAttribution': 'count'})

x = sum_trues_PA_df / counting_df

x = x.reset_index()

x_30 = x[:30]

x_30.columns
```
